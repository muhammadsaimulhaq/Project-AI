<div class="alert alert-block alert-alert">
    4.1. <b>Manage Null Values</b>
</div>

```python

df = df.drop('Year', axis=1)
```

```python
df=df.drop('Selling_Price', axis=1)
```

```python
# For Object
def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Car_Name','Owner','Transmission','Fuel_Type']
fillnaObjectMode(columns)
```

```python
# For Int Type
def fillnaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Car_Name','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']
fillnaIntMode(columns)
```

```python
df.info()
```

<div class="alert alert-block alert-alert">
    4.2. <b>Manage Data Types</b>
</div>

```python
df['Present_Price'] = df['Present_Price'].astype('int64')
```

```python
# For Float 
def changeFloatToInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Present_Price']
changeFloatToInt(columns)
```

```python
df.info()
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data)) 
        df[i] = data_LE.transform(df[i])
        
columns = ['Car_Name','Owner','Transmission','Fuel_Type','Seller_Type']
convertObjToInt(columns)
```

<div class="alert alert-block alert-alert">
    1. <b>Train - Test Splitting</b>
</div>

```python
import pandas as pd
df = pd.read_csv('processed_data.csv')
# df.info()
df.columns
```

```python
X = df.drop('Fuel_Type', axis=1) 
y = df['Fuel_Type']                
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False) 
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

<div class="alert alert-block alert-alert">
    1. <b> Model Training</b>
</div>

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(train_X, train_y)

print(model_svc)
```

```python
model_pred_svc = model_svc.predict(test_X)
model_pred_svc
```

```python
from sklearn.metrics import accuracy_score
model_acc_svc = accuracy_score(test_y, model_pred_svc)
print(f"{model_acc_svc * 100:.1f}%")
```
